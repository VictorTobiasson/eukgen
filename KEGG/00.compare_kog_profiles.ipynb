```python
root = "/data/luojaa/kegg_db/kofamkoala/"
kofam_nseq = pd.read_csv(root + "nseq.kofam.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"profile", 1:"nseq"})
kofam_len = pd.read_csv(root + "leng.kofam.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"profile", 1:"len"}).drop("profile", axis = 1)
kofam_neff = pd.read_csv(root + "effn.kofam.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"profile", 1:"neff"}).drop("profile", axis = 1)

keggeuk_nseq = pd.read_csv(root + "nseq.euk.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"nseq"})
keggprok_nseq = pd.read_csv(root + "nseq.prok.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"nseq"})
keggeuk_len = pd.read_csv(root + "leng.euk.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"len"})
keggprok_len = pd.read_csv(root + "leng.prok.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"len"})
keggeuk_name = pd.read_csv(root + "name.euk.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"name"})
keggprok_name = pd.read_csv(root + "name.prok.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"name"})
keggeuk_neff = pd.read_csv(root + "neff.euk.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"neff"})
keggprok_neff = pd.read_csv(root + "neff.prok.keggdb.txt", sep = "  ", engine = "python", header = None).rename(columns = {0:"desc", 1:"neff"})
```

```python
keggeuk_summary = pd.concat([keggeuk_name, keggeuk_nseq, keggeuk_len, keggeuk_neff], axis = 1).drop("desc", axis = 1)
keggprok_summary = pd.concat([keggprok_name, keggprok_nseq, keggprok_len, keggprok_neff], axis = 1).drop("desc", axis = 1)
```

```python
keggeuk_summary["nseq_unfilt"] = keggeuk_summary["nseq"].apply(lambda x: int(x.split("sequences")[0].split(" ")[3]))
keggeuk_summary["nseq_filt"] = keggeuk_summary["nseq"].apply(lambda x: int(x.split("sequences")[0].split(" ")[0]))
keggeuk_summary["length"] = keggeuk_summary["len"].apply(lambda x: int(x.split(" ")[0]))
keggeuk_summary.drop(["nseq", "len"], axis = 1, inplace = True)

keggprok_summary["nseq_unfilt"] = keggprok_summary["nseq"].apply(lambda x: int(x.split("sequences")[0].split(" ")[3]))
keggprok_summary["nseq_filt"] = keggprok_summary["nseq"].apply(lambda x: int(x.split("sequences")[0].split(" ")[0]))
keggprok_summary["length"] = keggprok_summary["len"].apply(lambda x: int(x.split(" ")[0]))
keggprok_summary.drop(["nseq", "len"], axis = 1, inplace = True)
```

```python
kofam_summary = pd.concat([kofam_nseq, kofam_len, kofam_neff], axis = 1)
kofam_summary["KOG"] = kofam_summary["profile"].apply(lambda x: x.split("/")[1].split(".")[0])
kofam_summary.drop("profile", axis = 1, inplace = True)
```

```python
plotA, dataA = plot_cumsum_counts(kofam_summary["nseq"], x_min = 1, plot_type = "default", title = "kofam_nseq")
plotB, dataB = plot_cumsum_counts(keggeuk_summary["nseq_unfilt"], x_min = 1, plot_type = "default", title = "keggeuk_nseq_filt")
plotC, dataC = plot_cumsum_counts(keggprok_summary["nseq_unfilt"], x_min = 1, plot_type = "default", title = "keggprok_nseq_filt")

unfiltered_data = pd.concat([dataA, dataB, dataC])
plot, data = plot_cumsum_counts(unfiltered_data, formatted_data=True, plot_type='default', title = "our profiles max out at 250 input seqs")
plot
```
