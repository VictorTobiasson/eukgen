### look at coverages of 60 vs 80% when scraping the eukaryotic search results
### compare distributions of hits per KOG target

```python
KOGS_w_both = "/data/luojaa/uid_stats/KOGs_w_both.csv"
relevant_kogs = set(pd.read_csv(KOGS_w_both)["KOG"])
```

```python
hhresults_euk_tsv = "/data/luojaa/kegg_db/search_euk_seq/kegg_euk_seq.cut.tsv"
hhresults_eukdf = pd.read_csv(hhresults_euk_tsv, sep = "\t")
hhresults_euk_knum = hhresults_eukdf[hhresults_eukdf["Query"] != hhresults_eukdf["Target"]]
accessions_euk = hhresults_euk_knum[hhresults_euk_knum["Target"].isin(relevant_kogs)]
```

```python
hhresults_euk60_tsv = "/data/luojaa/kegg_db/search_euk_seq/kegg_euk_seq.60.cut.tsv"
hhresults_euk60df = pd.read_csv(hhresults_euk60_tsv, sep = "\t")
hhresults_euk60_knum = hhresults_euk60df[hhresults_euk60df["Query"] != hhresults_euk60df["Target"]]
accessions_euk60 = hhresults_euk60_knum[hhresults_euk60_knum["Target"].isin(relevant_kogs)]
```

```python
all_domains_counts_euk = hhresults_euk_knum["Target"].value_counts()
all_domains_counts_euk60 = hhresults_euk60_knum["Target"].value_counts()
```

```python
plotA, dataA = plot_cumsum_counts(all_domains_counts_euk, x_min = 1, plot_type = "default", title = "euk proteins per KOG, 80% coverage hits")
plotB, dataB = plot_cumsum_counts(all_domains_counts_euk60, x_min = 1, plot_type = "default", title = "euk proteins per KOG, 60% coverage hits")
unfiltered_data = pd.concat([dataA, dataB])
plotC, dataC = plot_cumsum_counts(unfiltered_data, formatted_data=True, title='impact of coverage filters on number of eukaryotic hits', plot_type='default', x_label = "Number of euk proteins in microcosm")
plotC
```

```python
both_domains_counts_euk = accessions_euk["Target"].value_counts()
both_domains_counts_euk60 = accessions_euk60["Target"].value_counts()
```

```python
plot1, data1 = plot_cumsum_counts(both_domains_counts_euk, x_min = 1, plot_type = "default", title = "euk proteins per microcosm, 80% coverage hits")
plot2, data2 = plot_cumsum_counts(both_domains_counts_euk60, x_min = 1, plot_type = "default", title = "euk proteins per microcosm, 60% coverage hits")
```

```python
all_data = pd.concat([data1, data2])
plot, data = plot_cumsum_counts(all_data, formatted_data=True, title='impact of coverage filters on number of eukaryotic hits per microcosm', plot_type='default', x_label = "Number of euk proteins in microcosm")
plot
```
