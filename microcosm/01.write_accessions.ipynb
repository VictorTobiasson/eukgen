```python
import pandas as pd
import os
```

```python
kogs_df_searchable = "/data/luojaa/uid_stats/kogs_df_searchable.tsv"
kogs_df = pd.read_csv(kogs_df_searchable, sep = "\t")
```

```python
KOGS_w_both = "/data/luojaa/uid_stats/KOGs_w_both.csv"
relevant_kogs = pd.read_csv(KOGS_w_both)["KOG"]
```

```python
accessions_df = kogs_df.loc[:,["KOGID", "UID", "ISEUK", "SPECIES"]]
accessions_df = accessions_df[accessions_df["SPECIES"] != "none"] # some UIDs aren't in our sequence database
```

```python
accessions_df
```

```python
accessions_df = accessions_df[accessions_df["KOGID"].isin(relevant_kogs)]
```

```python
acc_counts = accessions_df.groupby(["KOGID", "ISEUK"]).count().reset_index()
superrelkogs = set(acc_counts[(acc_counts["ISEUK"] == True) & (acc_counts["UID"] > 5)]["KOGID"])
```

```python
accessions_outdf = accessions_df.groupby(["KOGID", "ISEUK"])["UID"].apply(lambda x: "\n".join(x)).reset_index()
```

```python
accessions_outdf = accessions_outdf[accessions_outdf["KOGID"].isin(superrelkogs)]
```

```python
outpathroot = "/data/luojaa/eukgen/processing/microcosm_backup/"
```

```python
kogids = list(accessions_outdf["KOGID"])
iseuks = list(accessions_outdf["ISEUK"])
uids = list(accessions_outdf["UID"])
```

```python
with open("/data/luojaa/eukgen/mmseqs/ran/bothdomains.ids", "w") as outfile:
    for i in range(len(accessions_outdf)):
        uidchunk = uids[i]
        for uid in uidchunk.split("\n"):
            print(uid,file=outfile)
```

```python
accessions_outdf
```

```python
for i in range(len(accessions_outdf)):
    kogid, iseuk, uid = kogids[i], iseuks[i], uids[i]
    os.system(f"mkdir {outpathroot}{kogid}")
    if iseuk:
        outpath = f"{outpathroot}{kogid}/{kogid}.query.acc"
        with open(outpath, "w") as outfile:
            print(uid, file=outfile)
    else:
        outpath = f"{outpathroot}{kogid}/{kogid}.target.acc"
        with open(outpath, "w") as outfile:
            print(uid, file=outfile)
```
